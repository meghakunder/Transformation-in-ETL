**Transformation in ETL**

**Question 1 : Define Data Transformation in ETL and explain why it is important.**

**Ans:**

Data Transformation is the process of converting raw extracted data into a clean, consistent, and usable format before loading it into the target system in the ETL process. It involves activities such as cleaning data, handling missing values, removing duplicates, changing data types, and standardizing formats.

Data Transformation is important because data collected from different sources is often incomplete, inconsistent, or unstructured. Without transformation, this data cannot be reliably used for analysis or reporting. Proper transformation improves data quality, ensures accuracy and consistency, and makes the data suitable for business intelligence and decision-making.

**Question 2 : List any four common activities involved in Data Cleaning.**

**Ans:**


* Handling missing values by replacing NULL values with appropriate defaults or removing incomplete records.

* Removing duplicate records to ensure only unique data is stored.

* Correcting inconsistent data formats, such as standardizing date or text formats.

* Validating data values to identify and fix incorrect or invalid entries.

**Question 3 : What is the difference between Normalization and Standardization?**

**Ans:**

Normalization is a data transformation technique used to scale numeric values into a specific range, usually between 0 and 1. It is mainly used when features have different ranges and is commonly applied in machine learning and analytics.

Standardization is a technique that transforms data so that it has a mean of 0 and a standard deviation of 1. It is useful when data needs to follow a standard normal distribution and when outliers are present.

In short, normalization focuses on rescaling data to a fixed range, while standardization focuses on centering and spreading data based on statistical measures.

**Question 4 : A dataset has missing values in the “Age” column. Suggest two techniques to handle this and
explain when they should be used.**

**Ans:**

1. Replacing missing values with the average (mean) age
This technique is used when most age values are present and the data is fairly balanced. Replacing missing values with the mean helps maintain the overall distribution of the data and is commonly used in analysis and modeling.

2. Removing records with missing age values
This technique is used when only a small number of records have missing age values and removing them does not significantly affect the dataset size or analysis results.

**Question 7 : Explain the difference between Data Integration and Data Mapping in ETL.**

**Ans:**

Data Integration is the process of combining data from multiple sources such as databases, files, or APIs into a single unified system. Its main purpose is to bring all related data together so it can be analyzed or reported in one place.

Data Mapping is the process of defining how fields from a source system match fields in a target system. It specifies which source column goes to which target column and what transformations are applied during the transfer.

In simple terms, data integration focuses on bringing data together, while data mapping focuses on how individual data fields are matched and transferred between source and target systems.

**Question 8 : Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.**

**Ans:**

When outliers exist in data, Z-score Standardization is preferred because it reduces the impact of extreme values. Min-Max Scaling is highly affected by outliers since it depends on minimum and maximum values.

Example Dataset
50, 55, 60, 65, 1000

Here, 1000 is an outlier.

Min Max Sacling

| Value | Min-Max Scaled |
| ----- | -------------- |
| 50    | 0              |
| 55    | 0.005          |
| 60    | 0.010          |
| 65    | 0.016          |
| 1000  | 1              |

Most values are pushed very close to 0 because the outlier (1000) increases the range.

Z-score Standardization Example

| Value | Z-score (approx) |
| ----- | ---------------- |
| 50    | -0.54            |
| 55    | -0.52            |
| 60    | -0.50            |
| 65    | -0.48            |
| 1000  | 2.04             |

Normal values remain reasonably spaced, and the outlier clearly stands out without compressing the rest.

Min-Max Scaling is sensitive to outliers because it depends on extreme values. Z-score Standardization is preferred when outliers exist as it maintains better distribution of normal data.
